**Learner Context**: New to web programming, comfortable with Python, prefers learning through discussion and exploring concepts.

**Chapters Covered**: "Hypermedia: A Reintroduction", "Components of a Hypermedia System", "A Web 1.0 Application", "Extending HTML As Hypermedia", "Htmx Patterns", "More Htmx Patterns", "A Dynamic Archive UI", "Tricks Of The Htmx Masters", and "Client Side Scripting" from "Hypermedia Systems"

**Key Concepts Discussed**:

1. **Hypermedia Definition**: Media with non-linear branching (e.g., hyperlinks). A **hypermedia control** encodes interaction information directly and completely within itself ‚Äî self-describing, no external knowledge needed. Think about the *whole system*: HTML + HTTP + browser + server working together.

2. **HTML's Two Hypermedia Controls**: Anchor tags (navigation, GET) and form tags (data submission, GET/POST). Forms enabled the web to become an *application* platform, not just documents. GET appends data to URL (bookmarkable); POST sends data in body. POST-Redirect-GET pattern prevents duplicate submissions.

3. **SPAs vs Hypermedia**: SPAs exchange JSON data, require client-side code with intimate knowledge of API structure (**tight coupling**). Hypermedia exchanges HTML with embedded controls ‚Äî server can change without breaking client (**loose coupling**). SPAs shift code/complexity to the browser.

4. **Why SPAs Won**: HTML froze as hypermedia in the mid-90s; user experience demands grew; JavaScript filled the gap. Won on *user experience*, not architectural superiority.

5. **Hypermedia-Driven Applications (HDAs)**: Apps using hypermedia exchanges as primary server communication. Libraries like **htmx** extend HTML's capabilities with declarative attributes (`hx-get`, `hx-target`). **Important**: browsers don't natively understand these attributes ‚Äî the htmx library scans for them, attaches behavior, handles requests/swaps; the browser just renders the resulting HTML.

6. **Transitional vs Hypermedia-First**: Rich Harris's "transitional" approach = SPA as home base with MPA sprinkled in. The book's approach = **hypermedia as default**, SPA only when genuinely needed. The mindset is flipped.

7. **When to Use What**: Plain HTML ‚Üí many apps. HTML + htmx ‚Üí most interactivity needs. SPA ‚Üí only when required (e.g., spreadsheets with cascading updates on every keystroke). **Can mix**: hypermedia for simple parts, SPA for complex features ‚Äî save your complexity budget.

8. **Browser Benefits "For Free"**: Back button, deep linking, caching ‚Äî all work naturally with hypermedia; must be manually reimplemented in SPAs.

9. **Four Components of a Hypermedia System**: Hypermedia (HTML), network protocol (HTTP), hypermedia server (responds to requests), hypermedia client (browser interprets responses). All four must work together.

10. **URLs**: Uniform Resource Locators uniquely identify resources. Components: scheme (protocol), host (domain), port, path, query, fragment. Relative URLs inherit from current document.

11. **HTTP Basics**: Simple request/response protocol. Request headers (like `Accept`) tell server preferences; response headers provide metadata. Content negotiation allows same URL to return different representations.

12. **HTTP Methods**: GET (read), POST (create/catch-all), PUT (replace), PATCH (partial update), DELETE (remove). Map roughly to CRUD. HTML only supports GET and POST natively ‚Äî a limitation htmx addresses. Methods are conventions, not enforced constraints.

13. **HTTP Response Codes**: 1xx informational, 2xx success, 3xx redirect, 4xx client error, 5xx server error. Subtle differences matter (302 vs 303). Using codes properly goes "with the grain" of the web.

14. **Caching**: `Cache-Control` header indicates cacheability. `Vary` header tells caches which request headers affect the response (important for htmx's `HX-Request` header).

15. **Hypermedia Servers**: Any language works ‚Äî hypermedia frees you from JavaScript lock-in. "HOWL" = Hypermedia On Whatever you'd Like.

16. **Hypermedia Clients**: Must generically interpret hypermedia controls. Browsers don't know what a "contact" is ‚Äî they just render HTML. JSON API clients expect fixed formats, which is why HATEOAS hasn't caught on there.

17. **REST (Original Meaning)**: Fielding's dissertation described the pre-API web. Most "REST APIs" today aren't RESTful by his definition.

18. **REST Constraints**: *Client-Server*: Obvious for web apps. *Statelessness*: Each request self-contained (sessions/cookies technically violate this ‚Äî pragmatic trade-off). *Caching*: Responses indicate cacheability. *Uniform Interface*: The core innovation. *Layered System*: Intermediaries (CDNs, proxies) allowed. *Code-On-Demand*: Scripting is optional ‚Äî should augment, not replace hypermedia.

19. **Uniform Interface Sub-constraints**: *Identification of resources*: URLs uniquely identify resources. *Manipulation through representations*: You exchange representations, not resources directly. *Self-descriptive messages*: All info needed is in the response itself. *HATEOAS*: Hypermedia As The Engine of Application State ‚Äî available actions encoded in HTML, not in client code.

20. **HATEOAS & API Churn**: Hypermedia APIs don't need versioning. Server changes, client automatically gets new controls. JSON APIs require documentation exchange and coordinated updates.

21. **Core Trade-off**: Hypermedia is less efficient (more bytes) but vastly more flexible ‚Äî clients need zero domain knowledge, just render what arrives.

22. **Routes & Handlers**: URLs map to Python functions via decorators (`@app.route("/contacts")`). The handler processes the request and returns a response (HTML or redirect).

23. **Same Path, Different Methods**: A single URL can have multiple handlers distinguished by HTTP method. `/contacts/new` uses GET to show a form, POST to process submission.

24. **Server-Side Rendering (SSR)**: Templates generate complete HTML on the server. The browser receives ready-to-render hypermedia, not JSON + client code.

25. **Template Inheritance**: Base layouts (`layout.html`) provide common "chrome"; child templates inject content into `{% block %}` sections. DRY on the server.

26. **Post/Redirect/Get (PRG)**: After a successful POST, redirect to a GET. This makes the "current page" a safe GET request, so refresh doesn't resubmit the form.

27. **Query Parameters vs Path Parameters**: Query: `?q=joe` for search/filtering. Path: `/contacts/42` for identifying specific resources.

28. **Forms as Hypermedia Controls**: `<form action="/contacts/new" method="post">` encodes everything the browser needs ‚Äî destination, method, input names. No client-side logic required.

29. **Error Handling Pattern**: If validation fails, re-render the same form with the contact object (which now contains errors). User sees their input preserved + error messages.

30. **Resource-Oriented URLs**: `/contacts` ‚Äî collection. `/contacts/new` ‚Äî form to create. `/contacts/42` ‚Äî single resource. `/contacts/42/edit` ‚Äî form to edit. `/contacts/42/delete` ‚Äî delete action.

31. **HTML's Method Limitations**: Forms only support GET and POST. No PUT, PATCH, or DELETE natively. This is a real constraint that htmx addresses.

32. **Accidentally RESTful**: By using hypermedia (HTML), the app naturally satisfies REST constraints ‚Äî HATEOAS, self-descriptive messages, loose coupling ‚Äî without explicit effort.

33. **Server-Side Factoring**: Unlike SPAs with tiny client-side components, hypermedia apps factor at coarser grain ‚Äî shared template sections rather than isolated widgets.

34. **The Web 1.0 Trade-off**: The app is architecturally sound and truly RESTful, but "clunky" ‚Äî full page reloads, flicker, lost scroll state.

35. **Four Limitations of HTML as Hypermedia**: Only anchors & forms can make requests. Only click/submit events trigger requests. Only GET & POST methods available. Only full page replacement possible.

36. **htmx Installation**: Simple script tag inclusion ‚Äî no build system required. Can use CDN with integrity hash for security, or download locally.

37. **htmx Core Philosophy**: Extends HTML declaratively via attributes. No JavaScript writing required by the developer.

38. **Request Attributes**: `hx-get`, `hx-post`, `hx-put`, `hx-patch`, `hx-delete` ‚Äî any element can issue any HTTP method. Value is the URL.

39. **Responses are HTML**: htmx expects HTML back, not JSON. Partial HTML snippets are typical.

40. **Targeting with `hx-target`**: CSS selector specifying where response HTML goes. Default is inside the triggering element.

41. **Swap Styles with `hx-swap`**: Controls *how* content is inserted. Common values: `innerHTML` (default), `outerHTML`, `beforeend`, `afterbegin`.

42. **Triggers with `hx-trigger`**: Specifies which event fires the request. Defaults: inputs/textareas/selects ‚Üí `change`, forms ‚Üí `submit`, everything else ‚Üí `click`.

43. **Event Filters**: Square bracket syntax filters events: `hx-trigger="keyup[ctrlKey && key == 'l']"`. Filter is JavaScript evaluated against the event object.

44. **Listening to Other Elements**: `from:` modifier listens elsewhere: `hx-trigger="keyup from:body"`.

45. **Event Bubbling**: Events bubble up through ancestors. You can catch descendant events without `from:`. Use `from:` when source is outside your element's ancestry.

46. **Passing Data ‚Äî Three Methods**: *Enclosing forms*: Auto-included for POST/PUT/PATCH/DELETE, NOT for GET. *`hx-include`*: Explicit CSS selector for inputs. *`hx-vals`*: Static or dynamic values.

47. **Why GET Doesn't Auto-Include Form Data**: htmx avoids URL clutter in browser history. Opt-in via `hx-include`.

48. **`id` vs `name` Attributes**: `id` is for browser-side targeting (CSS, htmx). `name` is for server-side data keys. Both can coexist.

49. **Relative CSS Selectors**: `closest`, `next`, `previous`, `find`, `this` ‚Äî select elements relative to current element.

50. **History Support with `hx-push-url`**: Pushes URL to browser history. Back/forward navigation works.

51. **The Eight Core Attributes**: `hx-get`, `hx-post`, `hx-put`, `hx-patch`, `hx-delete`, `hx-trigger`, `hx-target`, `hx-swap` ‚Äî address all four HTML limitations.

52. **htmx Trade-off**: Less "infrastructure" than SPA frameworks (no built-in state management, routing, component systems). Philosophy: stay simple.

53. **DOM (Document Object Model)**: Browser's internal tree representation of parsed HTML. Live and mutable ‚Äî enables partial updates without full page refresh.

54. **Transclusion**: Ted Nelson's 1980 concept ‚Äî including content into an existing document via hypermedia reference. htmx delivers this for HTML.

55. **`hx-boost`**: Converts all anchor and form elements within an element into AJAX-powered controls. Place on `<body>` to boost entire app.

56. **`hx-boost` Limitation**: New `<head>` content in responses is ignored (except `<title>`). Works best when all pages share the same layout.

57. **Attribute Inheritance**: Most htmx attributes cascade to children (like CSS). Override with explicit values on specific elements.

58. **Progressive Enhancement**: `hx-boost` degrades gracefully ‚Äî if JS disabled, links/forms work normally.

59. **Explicit vs Implicit Philosophy**: htmx generally prefers explicit attribute declarations over magic. `hx-boost` is the exception.

60. **Delete with HTTP DELETE**: Use `hx-delete="/contacts/{{ id }}"` on a button ‚Äî no form wrapper needed.

61. **Response Code 303**: After DELETE (or PUT/PATCH), use `redirect("/path", 303)` not default 302. Explicitly means "redirect and switch to GET."

62. **`hx-confirm`**: Add confirmation dialogs with `hx-confirm="Are you sure?"`. Uses browser's native `confirm()`.

63. **Graceful Degradation Trade-off**: The DELETE button won't work without JS. Supporting both htmx and fallback requires duplicate routes/handlers.

64. **Client-Side vs Server-Side Validation**: HTML5 `type="email"` handles format validation client-side (but can't be trusted). Uniqueness validation *requires* server.

65. **Inline Validation Pattern**: Add `hx-get`, `hx-target`, and `hx-trigger` to an input. On `change` event, htmx requests server validation and updates error span.

66. **Triggering Element Includes Its Own Value**: When an input triggers a request, its own name/value is automatically included.

67. **`keyup` Trigger for Live Validation**: Add `hx-trigger="change, keyup"` for validation as user types.

68. **Debouncing with `delay`**: Use `hx-trigger="change, keyup delay:200ms"` to wait until user pauses. Resets timer on each keystroke.

69. **`changed` Modifier**: Add `changed` to trigger to only fire when input value actually changes.

70. **Paging Basics**: Pass `page` query parameter, server returns appropriate slice.

71. **Click To Load Pattern**: Replace "Next" link with button using `hx-get`, `hx-target="closest tr"`, `hx-swap="outerHTML"`, `hx-select="tbody > tr"`.

72. **`hx-select`**: CSS selector to extract specific content from server response before swapping.

73. **Infinite Scroll Pattern**: Same as click-to-load but trigger on `revealed` event instead of click.

74. **Modals Caution**: Modal windows introduce client-side state that doesn't map to URLs. Use sparingly.

75. **`display: none` Accessibility**: Hides elements from accessibility tree and keyboard focus, not just visually.

76. **The Core htmx Pattern**: "Event ‚Üí Request ‚Üí Swap." Many UX patterns become variations on this theme.

77. **Active Search Pattern**: Add `hx-get`, `hx-trigger="search, keyup delay:200ms changed"`, and `hx-target` to a search input.

78. **The `search` Event**: A non-standard event triggered when user clears search input or hits enter.

79. **Template Factoring Pattern**: Break templates into reusable fragments. Same fragment serves full page and htmx partial updates.

80. **Jinja `include` Directive**: `{% include 'rows.html' %}` inserts another template's content at that position.

81. **htmx Request Headers**: `HX-Request`, `HX-Trigger`, `HX-Target`, `HX-Boosted`, `HX-Current-URL` automatically added.

82. **Using `HX-Trigger` for Conditional Responses**: Server checks which element initiated the request. Allows same endpoint to return different content.

83. **Bookmarks and `HX-Trigger`**: When user opens bookmarked URL directly, there's no htmx ‚Äî no `HX-Trigger` header. Server returns full page.

84. **`Vary` Header for Caching**: When returning different content based on htmx headers, add `Vary: HX-Trigger` to responses.

85. **`hx-push-url` with Active Search**: Updates browser URL as user types, enabling bookmarks and back button.

86. **Request Indicators with `hx-indicator`**: CSS selector pointing to a spinner element to show during requests.

87. **Indicators Inside Elements**: For one-time requests, place indicator *inside* target element ‚Äî self-cleaning.

88. **Throttling for Testing**: Use browser dev tools to simulate slow networks.

89. **Lazy Loading Pattern**: Defer expensive operations to secondary requests. Main page loads fast.

90. **`hx-trigger="load"`**: Synthetic htmx event fired when element is added to DOM.

91. **`hx-trigger="revealed"`**: Synthetic htmx event fired when element scrolls into viewport.

92. **Synthetic vs Native Events**: `load` and `revealed` are htmx-invented events with no meaning in vanilla JavaScript.

93. **Lazy Loading Sequence**: Main page ‚Üí placeholder renders ‚Üí htmx fires secondary request ‚Üí response fills placeholder.

94. **Perceived Performance**: Lazy loading improves *perceived* performance ‚Äî users can interact immediately.

95. **Inline Delete Pattern**: Add `hx-delete`, `hx-target="closest tr"`, `hx-swap="outerHTML"`, and `hx-confirm` to a delete link.

96. **Empty Response for Removal**: When `hx-swap="outerHTML"` receives empty string, target element is removed from DOM.

97. **Distinguishing Delete Sources**: Add `id` to delete buttons to differentiate via `HX-Trigger` header.

98. **htmx Swapping Model**: Content goes through phases: `htmx-swapping` class ‚Üí swap delay ‚Üí `htmx-settling` class ‚Üí settle delay. Enables CSS transitions.

99. **CSS Transitions with htmx**: Write CSS rules targeting `htmx-swapping` or `htmx-settling` classes.

100. **Swap Delay Modifier**: `hx-swap="outerHTML swap:1s"` gives CSS transitions time to complete.

101. **Fade-Out Delete Pattern**: Combine swap delay with CSS transition on `htmx-swapping` class.

102. **Bulk Delete Pattern**: Wrap table and delete button in a `<form>`. Add checkboxes with same `name` attribute.

103. **Checkbox Form Behavior**: Multiple checkboxes with same `name` submit all checked values as a list.

104. **No Redirect for Same-Page Updates**: When already on the page being re-rendered, skip the redirect.

105. **Full-Stack Advantage**: Controlling both frontend and backend enables returning partial HTML based on request context.

106. **Long-Running Process Problem**: Blocking until complete = bad UX. Non-blocking with immediate response = better UX.

107. **Common Async UI Patterns**: Email when complete, manual refresh, auto-refresh with JS timer.

108. **Non-Blocking Design**: Server starts background job, responds immediately with progress UI. Client polls for updates.

109. **Polling**: Client repeatedly asks server "are you done yet?" at intervals.

110. **Fixed-Rate Polling**: `hx-trigger="every 3s"` ‚Äî polls indefinitely on a timer.

111. **Load Polling**: `hx-trigger="load delay:500ms"` ‚Äî each response triggers the next request. Server controls continuation.

112. **Load Polling Chain**: Response includes trigger ‚Üí element loads ‚Üí fires request after delay ‚Üí repeat. Omit trigger to stop.

113. **Same URL, Multiple Methods Pattern**: POST starts job; GET checks status; DELETE resets/cancels.

114. **Single Template, Multiple States**: One template handles all UI states with Jinja `{% if %}` / `{% elif %}` blocks.

115. **Archiver API Pattern**: `status()`, `progress()`, `run()`, `reset()`, `archive_file()`. Clean separation of concerns.

116. **CSS Progress Bar**: Outer div (frame) + inner div (bar). Set inner div's `width` as percentage.

117. **ARIA for Custom Widgets**: `role="progressbar"` and `aria-valuenow="50"` make div-based progress bars accessible.

118. **File Downloads via HTTP**: `Content-Disposition: attachment` header tells browser to save response as file.

119. **`hx-boost="false"` for Downloads**: Opt out of AJAX for file download links.

120. **CSS Transitions Don't Fire on Replacement**: Transitions animate property *changes* on existing elements.

121. **htmx Settling Mechanism**: htmx matches elements by `id`, copies old attributes, swaps, then restores new attributes ‚Äî creating animatable change.

122. **Stable ID Requirement**: For settling/smooth transitions, element needs the **same** `id` in both old and new HTML.

123. **Hypermedia-Friendly Scripting**: Scripting is acceptable when it *enhances* rather than *replaces* hypermedia.

124. **HTML Entities**: `&downarrow;` ‚Üí ‚Üì, `&amp;` ‚Üí &, `&lt;` ‚Üí <, `&copy;` ‚Üí ¬©.

125. **Flask Context Processors**: `@app.context_processor` injects variables into all templates automatically.

126. **Markdown Soup**: Using Markdown elements for *appearance* rather than semantic *meaning*. Prefer conformant HTML.

127. **`hx-swap` Modifiers Beyond Basics**: `settle` (delay before attribute settling), `show` (scroll element into viewport, e.g., `show:body:top`), `scroll` (scroll scrollable element to top/bottom), `focus-scroll` (scroll to focused element).

128. **`delay` vs `throttle`**: `delay:200ms` resets timer on every event, fires after silence (search-as-you-type). `throttle:200ms` fires immediately, then ignores events for interval (scroll/resize handlers).

129. **Additional `hx-trigger` Modifiers**: `once` (trigger only once), `target` (filter to events directly on specific element), `consume` (cancel event, prevent bubbling), `queue` (control event queuing during in-flight requests).

130. **Trigger Filters**: JavaScript expression in square brackets evaluated on each event: `hx-trigger="click[contactRetrievalEnabled()]"`. Request proceeds only if truthy.

131. **`intersect` Synthetic Event**: More fine-grained than `revealed` ‚Äî can specify threshold (e.g., trigger when 50% visible).

132. **`hx-preserve`**: Keeps element unchanged during swaps ‚Äî useful for preserving partially-filled forms or video players. Requires stable `id`.

133. **`hx-sync`**: Synchronize requests between multiple elements. Prevents race conditions. Value `this` on parent syncs all children.

134. **`hx-disable`**: Disables all htmx behavior on element and children. Security feature for user-generated content.

135. **Events as "Glue"**: DOM is essentially an event bus. Events connect DOM, HTML, htmx, and scripting.

136. **htmx-Emitted Events**: `htmx:load` (content loaded), `htmx:configRequest` (before request ‚Äî can modify/cancel), `htmx:afterRequest` (after completion), `htmx:abort` (send to element to cancel in-flight request).

137. **`htmx:configRequest` for Global Request Modification**: Listen on `document.body`, modify `evt.detail.headers` or `evt.detail.parameters` to add auth tokens to every request.

138. **`localStorage`**: Browser-side key-value storage that persists across sessions. Server cannot access directly ‚Äî JavaScript must explicitly read and send it. Different from cookies (auto-sent with every request).

139. **`detail` Property Convention**: Web standard for custom events ‚Äî data goes in `detail` to avoid conflicts with native event properties.

140. **`htmx:abort` Pattern**: Send this event to an htmx element to cancel its in-flight request. Useful for "Cancel" buttons.

141. **Hyperscript for UI State**: `_="on htmx:beforeRequest remove @disabled from me"` ‚Äî enables Cancel button only while request is in flight. `@disabled` grays out and prevents clicks.

142. **`HX-Trigger` Response Header**: Server can trigger client-side events by including event name in response header. Enables decoupled coordination.

143. **Decoupled Coordination Pattern**: Button announces event, doesn't know who listens. Table listens on `body` (events bubble up). Loose coupling.

144. **htmx Response Headers**: `HX-Location` (client-side redirect), `HX-Push-Url` (push URL to location bar), `HX-Refresh` (refresh page), `HX-Retarget` (change swap target).

145. **Special Response Codes**: `204 No Content` (htmx skips swap entirely), `286` (stops polling).

146. **`htmx:beforeSwap` Event**: Intercept before swap. Can check `evt.detail.xhr.status` to handle error codes. Can modify whether/where swap occurs.

147. **Three Approaches to "Update Other Content"**: (1) **Expand target** (simplest): Make target large enough. (2) **Out-of-band swaps**: Piggyback extra content with `hx-swap-oob="true"`. (3) **Server-triggered events**: Use `HX-Trigger` header (most flexible, most complex).

148. **Out-of-Band Swaps**: Response includes elements with `hx-swap-oob="true"` and matching `id`. htmx extracts these, swaps by ID, then swaps remaining content into normal target.

149. **Complexity Budget**: htmx saves complexity on simple parts, freeing budget for genuinely complex features.

150. **Debugging with `htmx.logAll()`**: Call in console or script to log all htmx internal events.

151. **Chrome `monitorEvents()`**: Console-only function to monitor all events on an element. Use `monitorEvents($0)` for currently selected element.

152. **Content Security Policy (CSP)**: Browser security feature restricting allowed scripts/sources. htmx works with most CSPs. Exception: disabling `eval()` breaks trigger filters.

153. **htmx Configuration via Meta Tag**: `<meta name="htmx-config" content='{"defaultSwapStyle":"outerHTML"}'>` ‚Äî override defaults globally.

154. **Conformant HTML**: Use elements for their specified purpose per the HTML spec, letting browsers and assistive tech extract meaning.

155. **Scripting is Allowed in Hypermedia**: The web wasn't just "for documents" ‚Äî hypermedia systems like HyperCard had scripting before the web. The limitation is HTML's underdeveloped implementation, not the concept.

156. **Two Constraints for Hypermedia-Friendly Scripting**: (1) Main data exchange must be hypermedia (HTML), not JSON. (2) Client-side state outside the DOM is kept to a minimum.

157. **Scripting's Sweet Spot**: Interaction design ‚Äî how users physically interact with UI (menus, keyboard shortcuts, animations). Business logic and presentation logic belong on the server.

158. **SPA JS Wisdom May Not Apply**: Conventional JavaScript best practices come from SPA culture and may not suit HDAs.

159. **VanillaJS**: Plain JavaScript with no frameworks. Tongue-in-cheek name; no installation needed ‚Äî just write JavaScript.

160. **What HDAs Don't Need**: Client-side routing, reactive templates, SSR for JS, hydration, network request abstraction. Hypermedia handles these, dramatically reducing framework needs.

161. **JavaScript Type Coercion**: `++` on a string like `"0"` converts to number, increments, converts back. Convenient but can be surprising.

162. **Separation of Concerns (SoC) Illusion**: Splitting HTML/CSS/JS into files doesn't truly separate concerns ‚Äî they remain tightly coupled. Changing HTML often breaks JS.

163. **jQuery Soup**: Disorganized JavaScript where behavior is hard to find, code reuse is difficult, and unrelated event handlers mix together.

164. **Locality of Behavior (LoB)**: "The behavior of a unit of code should be as obvious as possible by looking only at that unit of code." You should be able to tell what a button does by looking at it.

165. **LoB vs SoC**: LoB doesn't require implementation at the use site, just *invocation*. The code can live elsewhere as long as the element makes clear what it triggers.

166. **RSJS (Reasonable System for JavaScript Structure)**: Guidelines for vanilla JS ‚Äî use `data-` attributes to invoke behavior, one component per file named after the attribute.

167. **State in the DOM**: Instead of JavaScript variables (`let isOpen = false`), store state in DOM attributes (e.g., `hidden` attribute). If HTML is snapshotted and restored, state survives.

168. **`hidden` HTML Attribute**: Boolean attribute ‚Äî presence means hidden, absence means visible. `hidden="false"` still hides (boolean attributes are presence-based, not value-based).

169. **ARIA (Accessible Rich Internet Applications)**: W3C standard for making custom widgets accessible. Two components: `role` attribute (promises widget behavior) and `aria-*` attributes (states and relationships).

170. **"No ARIA is Better than Bad ARIA"**: Adding a `role` is a *promise* to implement that behavior. Breaking the promise makes UX worse, not better.

171. **Alpine.js**: JavaScript library for adding behavior directly in HTML via `x-` attributes. Modern jQuery replacement, supports reactivity.

172. **`x-data`**: Declares reactive state as a JavaScript object on an element. Properties accessible to all descendants.

173. **`x-text`**: Binds element's text content to a reactive variable. Updates automatically when variable changes.

174. **`x-on:click` (or `@click`)**: Event listener with modifiers like `.outside` (click outside element), `.throttle`, `.debounce`. Can listen for custom events.

175. **Reactivity**: When you change data, Alpine automatically updates the DOM. No manual `element.textContent = value` calls.

176. **`x-if` with `<template>`**: Conditionally shows/hides content. Browser doesn't render `<template>` by default; Alpine makes it appear when condition is true.

177. **`x-model`**: Two-way binding between form inputs and data. Checkbox checked ‚Üí value added to array; array modified ‚Üí checkbox updates.

178. **`$root`**: Alpine's special symbol for the nearest ancestor with `x-data`. Useful for referencing the component's container.

179. **`$el`**: Alpine's symbol for "this element" ‚Äî the element the script is on.

180. **`htmx.ajax()`**: htmx's JavaScript API for triggering hypermedia exchanges from code. Same hypermedia exchange, just invoked programmatically.

181. **_hyperscript**: Non-JavaScript scripting language created alongside htmx. English-like syntax, event-driven (not reactive), makes DOM manipulation easy.

182. **_hyperscript Syntax**: `on click increment the textContent of the previous <output/>` ‚Äî reads like English. `the` is optional for readability.

183. **_hyperscript `previous` and `next`**: Built-in keywords for finding neighboring elements. No equivalent in vanilla JS DOM API.

184. **_hyperscript `from` Clause**: Listen for events elsewhere while keeping code on the element it affects. `on keydown from window` ‚Äî global listener, local code.

185. **_hyperscript `me`**: Pronoun referring to the element the script is on.

186. **_hyperscript Async Transparency**: Async functions can be called as if synchronous. No `async/await` ceremony ‚Äî the runtime handles it.

187. **Web Components**: Custom Elements + Shadow DOM + `<template>`. If you want one feature, you get all three. Use when element "could reasonably be a built-in HTML element" (e.g., code editor).

188. **Best Libraries for HDAs**: (1) Mutate DOM but don't communicate via JSON. (2) Respect HTML norms (use `<input>` for values). (3) Emit many custom events.

189. **Events > Callbacks**: Libraries with rich event models integrate more naturally with htmx than callback-heavy libraries.

190. **Wrapper Pattern for Callbacks**: If a library uses callbacks (like SweetAlert2), wrap it in a function that emits events instead, bridging to the event-driven htmx world.

191. **`dispatchEvent()`**: Standard way to programmatically fire custom events. Any element can emit; htmx or other listeners can respond.

192. **Three Pragmatic Guidelines**: (1) Avoid JSON APIs. (2) Avoid state outside the DOM. (3) Favor events over callbacks. Break if needed, but don't make it a habit.

193. **Pragmatic Tool Selection**: VanillaJS, Alpine.js, or _hyperscript ‚Äî each has strengths. Can even mix within an app. Whatever feels right is probably right enough.

So far we have been focusing on using hypermedia to build Hypermedia-Driven Applications. In doing so we are following and taking advantage of the native network architecture of the web, and building a RESTful system, in the original sense of that term.

However, today, we should acknowledge that many web applications are often _not_ built using this approach. Instead, they use a Single Page Application front end library such as React to build their application, and they interact with the server via a JSON API. This JSON API almost never uses hypermedia concepts. Rather JSON APIs tend to be _Data APIs_ , that is, an API that simply returns structured domain data to the client without any hypermedia control information. The client itself must know how to interpret the JSON Data: what end points are associated with the data, how certain fields should be interpreted, and so on.

Now, believe it or not, we _have_ been creating an API for Contact.app.

This may sound confusing to you: an API? We have just been creating a web application, with handlers that just return HTML.

How is that an API?

It turns out that Contact.app is, indeed, providing an API. It just happens to be a _hypermedia_ API that a _hypermedia client_ , that is, a browser, understands. ## Hypermedia APIs & JSON Data APIs

So, we have a hypermedia API for Contact.app. Should we include a Data API for Contact.app as well?

Sure! The existence of a hypermedia API _in no way means_ that you can‚Äôt _also_ have a Data API. In fact, this is a common situation in traditional web applications: there is the ‚Äúweb application‚Äù that is entered through that entry point URL, say `https://mywebapp.example.com/`. And there is also a _separate_ JSON API that is accessible through another URL, perhaps `https://api.mywebapp.example.com/v1`.

This is a perfectly reasonable way to split up the hypermedia interface to your application and the Data API you provide to other, non-hypermedia clients.

Why would you want to include a Data API along with a hypermedia API? Well, because _non-hypermedia clients_ might also want to interact with your application as well.

For example:

  * Perhaps you have a mobile application that isn‚Äôt built using Hyperview. That application will need to interact with your server somehow, and using the existing HTML API would almost certainly be a poor fit! You want programmatic access to your system via a Data API, and JSON is a natural choice for this.

  * Perhaps you have an automated script that needs to interact with the system on a regular basis. For example, maybe we have a bulk-import job that runs nightly, and needs to import/sync thousands of contacts. While it would be possible to script this against the HTML API, it would also be annoying: parsing HTML in scripts is error prone and tedious. It would be better to have a simple JSON API for this use case.

  * Perhaps there are 3rd party clients who wish to integrate with your system‚Äôs data in some way. Maybe a partner wants to synchronize data nightly. As with the bulk-import example, this isn‚Äôt a great use case for an HTML-based API, and it would make more sense to provide something more amenable to scripting.




For all of these use cases, a JSON Data API makes sense: in each case the API is not being consumed by a hypermedia client, so presenting an HTML-based hypermedia API would be inefficient and complicated for the client to deal with. A simple JSON Data API fits the bill for what we want and, as always, we recommend using the right tool for the job.

### Differences Between Hypermedia APIs & Data APIs

Let‚Äôs accept for a moment that we _are_ going to have a Data API for our application, in addition to our hypermedia API. At this point, some developers may be wondering: why have both? Why not have a single API, the JSON Data API, and have multiple clients use this one API to communicate with it?

Isn‚Äôt it redundant to have both types of APIs for our application?

This is a reasonable point: we do advocate having multiple APIs to your web application if necessary and, yes, this may lead to some redundancy in code. However, there are distinct advantages to both sorts of APIs and, even more so, distinct requirements for both sorts of APIs.

By supporting both of these types of APIs separately you can get the strengths of both, while keeping their varying styles of code and infrastructure needs cleanly split out.

Let‚Äôs contrast the needs of JSON APIs with Hypermedia APIs:

JSON API Needs | Hypermedia API  
---|---  
It must remain stable over time: you cannot change the API willy-nilly or you risk breaking clients that use the API and expect certain end points to behave in certain ways. | There is no need to remain stable over time: all URLs are discovered via HTML responses, so you can be much more aggressive in changing the shape of a hypermedia API.  
It must be versioned: related to the first point, when you do make a major change, you need to version the API so that clients that are using the old API continue to work. | Versioning is not an issue, another strength of the hypermedia approach.  
It should be rate limited: since data APIs are often used by other clients, not just your own internal web application, requests should be rate limited, often by user, in order to avoid a single client overloading the system. | Rate limiting probably isn‚Äôt as important beyond the prevention of Distributed Denial of Service (DDoS) attacks.  
It should be a general API: since the API is for _all_ clients, not just for your web application, you should avoid specialized end points that are driven by your own application needs. Instead, the API should be general and expressive enough to satisfy as many potential client needs as possible. | The API can be _very specific_ to your application needs: since it is designed only for your particular web application, and since the API is discovered through hypermedia, you can add and remove highly tuned end points for specific features or optimization needs in your application.  
Authentication for these sorts of API is typically token based, which we will discuss in more detail later. | Authentication is typically managed through a session cookie established by a login page.  
  
These two different types of APIs have different strengths and needs, so it makes sense to use both. The hypermedia approach can be used for your web application, allowing you to specialize the API for the ‚Äúshape‚Äù of your application. The Data API approach can be used for other, non-hypermedia clients like mobile, integration partners, etc.

Note that by splitting these two APIs apart, you reduce the pressure to constantly change a general Data API to address application needs. Your Data API can focus on remaining stable and reliable, rather than requiring a new version with every added feature.

This is the key advantage of splitting your Data API from your Hypermedia API.

## Adding a JSON Data API To Contact.app

Alright, so how are we going to add a JSON Data API to our application? One approach, popularized by the Ruby on Rails web framework, is to use the same URL endpoints as your hypermedia application, but use the HTTP `Accept` header to determine if the client wants a JSON representation or an HTML representation. The HTTP `Accept` header allows a client to specify what sort of Multipurpose Internet Mail Extensions (MIME) types, that is file types, it wants back from the server: JSON, HTML, text and so on.

So, if the client wanted a JSON representation of all contacts, they might issue a `GET` request that looks like this:
```
Accept: application/json

GET /contacts

```

A request for a JSON representation of all contacts

If we adopted this pattern then our request handler for `/contacts/` would need to be updated to inspect this header and, depending on the value, return a JSON rather than HTML representation for the contacts. Ruby on Rails has support for this pattern baked into the framework, making it very easy to switch on the requested MIME type.

Unfortunately, our experience with this pattern has not been great, for reasons that should be clear given the differences we outlined between Data and hypermedia APIs: they have different needs and often take on very different ‚Äúshapes‚Äù, and trying to pound them into the same set of URLs ends up creating a lot of tension in the application code.

Given the different needs of the two APIs and our experience managing multiple APIs like this, we think separating the two, and, therefore, breaking the JSON Data API out to its own set of URLs is the right choice. This will allow us to evolve the two APIs separately from one another, and give us room to improve each independently, in a manner consistent with their own individual strengths.

### Picking a Root URL For Our API

Given that we are going to split our JSON Data API routes out from our regular hypermedia routes, where should we place them? One important consideration here is that we want to make sure that we can version our API cleanly in some way, regardless of the pattern we choose.

Looking around, a lot of places use a subdomain for their APIs, something like `https://api.mywebapp.example.com` and, in fact, often encode versioning in the subdomain: `https://v1.api.mywebapp.example.com`.

While this makes sense for large companies, it seems like a bit of overkill for our modest little Contact.app. Rather than using subdomains, which are a pain for local development, we will use sub-paths within the existing application:

  * We will use `/api` as the root for our Data API functionality

  * We will use `/api/v1` as the entry point for version 1 of our Data API




If and when we decide to bump the API version, we can move to `/api/v2` and so on.

This approach isn‚Äôt perfect, of course, but it will work for our simple application and can be adapted to a subdomain approach or various other methods at a later point, when our Contact.app has taken over the internet and we can afford a large team of API developers. :)

#### Authentication in web applications

In the HTML web application world, authentication has traditionally been done via a login page that asks a user for their username (often their email) and a password. This password is then checked against a database of (hashed) passwords to establish that the user is who they say they are. If the password is correct, then a _session cookie_ is established, indicating who the user is. This cookie is then sent with every request that the user makes to the web application, allowing the application to know which user is making a given request.

**HTTP Cookies** HTTP Cookies are kind of a strange feature of HTTP. In some ways they violate the goal of remaining stateless, a major component of the RESTful architecture: a server will often use a session cookie as an index into state kept on the server ‚Äúon the side‚Äù, such as a cache of the last action performed by the user. 

Nonetheless, cookies have proven extremely useful and so people tend not to complain about this aspect of them too much (We are not sure what our other options would be here!) An interesting example of pragmatism gone (relatively) right in web development.

In comparison with the standard web application approach to authentication, a JSON API will typically use some sort of _token based_ authentication: an authentication token will be established via a mechanism like OAuth, and that authentication token will then be passed, often as an HTTP Header, with every request that a client makes.

At a high level this is similar to what happens in normal web application authentication: a token is established somehow and then that token is part of every request. However, in practice, the mechanics tend to be wildly different:

  * Cookies are part of the HTTP specification and can be easily _set_ by an HTTP Server.

  * JSON Authentication tokens, in contrast, often require elaborate exchange mechanics like OAuth to be established.




These differing mechanics for establishing authentication are yet another good reason for splitting up our JSON and hypermedia APIs.

### The Model View Controller (MVC) Paradigm

One thing you may have noticed about the handlers for our JSON API is that they are relatively simple and regular. **The Model View Controller pattern**

The Model View Controller design pattern is a classic architectural pattern in software development, and was a major influence in early web development. It is no longer emphasized as heavily, as web development has split into frontend and backend camps, but most web developers are still familiar with the idea.

Traditionally, the MVC pattern mapped into web development like so:

  * Model - A collection of ‚Äúdomain‚Äù classes that implement all the logic and rules for the particular domain your application is designed for. The model typically provides ‚Äúresources‚Äù that are then presented to clients as HTML ‚Äúrepresentations.‚Äù

  * View - Typically views would be some sort of client-side templating system, and would render the aforementioned HTML representation for a given Model instance.

  * Controller - The controller‚Äôs job is to take HTTP requests, convert them into sensible requests to the Model and forward those requests on to the appropriate Model objects. It then passes the HTML representation back to the client as an HTTP response.

## HTML Notes: Microformats

[Microformats](https://microformats.org/) is a standard for embedding machine-readable structured data in HTML. It uses classes to mark certain elements as containing information to be extracted, with conventions for extracting common properties like name, URL and photo without classes. By adding these classes into the HTML representation of an object, we allow the properties of the object to be recovered from the HTML. For example, this simple HTML:
```
<a class="h-card" href="https://john.example">
  <img src="john.jpg" alt=""> John Doe
</a>
```

can be parsed into this JSON-like structure by a microformats parser:
```
{
  "type": ["h-card"],
  "properties": {
    "name": ["John Doe"],
    "photo": ["john.jpg"],
    "url": ["https://john.example"]
  }
}
```

Using a variety of properties and nested objects, we could mark up every bit of information about a contact, for example, in a machine-readable way.

As explained in the above chapter, trying to use the same mechanism for human and machine interaction is not a good idea. Your human-facing and machine-facing interfaces may end up being limited by each other. If you want to expose domain-specific data and actions to users and developers, a JSON API is a great option.

However, microformats are way easier to adopt. A protocol or standard that requires websites to implement a JSON API has a high technical barrier. In comparison, any website can be augmented with microformats simply by adding a few classes. Other HTML-embedded data formats like microdata, Open Graph are similarly easy to adopt. This makes microformats useful for cross-website (dare we say _web-scale_) systems like the [IndieWeb](https://indieweb.org), which uses it pervasively.

165. **LoB vs SoC**: LoB doesn't require implementation at the use site, just *invocation*. The code can live elsewhere as long as the element makes clear what it triggers.

166. **RSJS (Reasonable System for JavaScript Structure)**: Guidelines for vanilla JS ‚Äî use `data-` attributes to invoke behavior, one component per file named after the attribute.

167. **State in the DOM**: Instead of JavaScript variables (`let isOpen = false`), store state in DOM attributes (e.g., `hidden` attribute). If HTML is snapshotted and restored, state survives.

168. **`hidden` HTML Attribute**: Boolean attribute ‚Äî presence means hidden, absence means visible. `hidden="false"` still hides (boolean attributes are presence-based, not value-based).

169. **ARIA (Accessible Rich Internet Applications)**: W3C standard for making custom widgets accessible. Two components: `role` attribute (promises widget behavior) and `aria-*` attributes (states and relationships).

170. **"No ARIA is Better than Bad ARIA"**: Adding a `role` is a *promise* to implement that behavior. Breaking the promise makes UX worse, not better.

171. **Alpine.js**: JavaScript library for adding behavior directly in HTML via `x-` attributes. Modern jQuery replacement, supports reactivity.

172. **`x-data`**: Declares reactive state as a JavaScript object on an element. Properties accessible to all descendants.

173. **`x-text`**: Binds element's text content to a reactive variable. Updates automatically when variable changes.

174. **`x-on:click` (or `@click`)**: Event listener with modifiers like `.outside` (click outside element), `.throttle`, `.debounce`. Can listen for custom events.

175. **Reactivity**: When you change data, Alpine automatically updates the DOM. No manual `element.textContent = value` calls.

176. **`x-if` with `<template>`**: Conditionally shows/hides content. Browser doesn't render `<template>` by default; Alpine makes it appear when condition is true.

177. **`x-model`**: Two-way binding between form inputs and data. Checkbox checked ‚Üí value added to array; array modified ‚Üí checkbox updates.

178. **`$root`**: Alpine's special symbol for the nearest ancestor with `x-data`. Useful for referencing the component's container.

179. **`$el`**: Alpine's symbol for "this element" ‚Äî the element the script is on.

180. **`htmx.ajax()`**: htmx's JavaScript API for triggering hypermedia exchanges from code. Same hypermedia exchange, just invoked programmatically.

181. **_hyperscript**: Non-JavaScript scripting language created alongside htmx. English-like syntax, event-driven (not reactive), makes DOM manipulation easy.

182. **_hyperscript Syntax**: `on click increment the textContent of the previous <output/>` ‚Äî reads like English. `the` is optional for readability.

183. **_hyperscript `previous` and `next`**: Built-in keywords for finding neighboring elements. No equivalent in vanilla JS DOM API.

184. **_hyperscript `from` Clause**: Listen for events elsewhere while keeping code on the element it affects. `on keydown from window` ‚Äî global listener, local code.

185. **_hyperscript `me`**: Pronoun referring to the element the script is on.

186. **_hyperscript Async Transparency**: Async functions can be called as if synchronous. No `async/await` ceremony ‚Äî the runtime handles it.

187. **Web Components**: Custom Elements + Shadow DOM + `<template>`. If you want one feature, you get all three. Use when element "could reasonably be a built-in HTML element" (e.g., code editor).

188. **Best Libraries for HDAs**: (1) Mutate DOM but don't communicate via JSON. (2) Respect HTML norms (use `<input>` for values). (3) Emit many custom events.

189. **Events > Callbacks**: Libraries with rich event models integrate more naturally with htmx than callback-heavy libraries.

190. **Wrapper Pattern for Callbacks**: If a library uses callbacks (like SweetAlert2), wrap it in a function that emits events instead, bridging to the event-driven htmx world.
